```python
# Reading data
import pandas as pd
import numpy as np

df = pd.read_pickle("../data/preprocess_train.pkl")
df.head(30)
```

```python
# Assemble Embedding Vectors

import time
from gensim.models import FastText, KeyedVectors

start=time.time()
FastText_embedding = KeyedVectors.load_word2vec_format("../data_source/wiki.en.vec")
end = time.time()
print("Loading the embedding took %d seconds"%(end-start))
```

```python
def handle_out_of_vocab(embedding,in_txt):
    out = None
    for word in in_txt:
        try:
            tmp = embedding[word]
            tmp = tmp.reshape(1,len(tmp))
            
            if out is None:
                out = tmp
            else:
                out = np.concatenate((out,tmp),axis=0)
        except:
            pass
    
    return out
        

def assemble_embedding_vectors(data):
    out = None
    row_num = None
    numw = 0
    for item in data:
        tmp = handle_out_of_vocab(FastText_embedding,item)
        if tmp is not None:
            dim = tmp.shape[1]
            if out is not None:
                vec = np.mean(tmp,axis=0)
                vec = vec.reshape((1,dim))
                out = np.concatenate((out,vec),axis=0)
            else:
                out = np.mean(tmp,axis=0).reshape((1,dim))                                            
        else:
            #if row_num is not None:
            #    row_num = np.concatenate((row_num, numw), axis=0)
            #else:
            #    row_num = [numw]
            print(item)
            print(numw)

            pass
        numw += 1
        
        
    return out
```

```python
EmbeddingVectors = assemble_embedding_vectors(np.array(df["text_final"]))
```

```python
df_new = df.drop([19, 24, 4234, 5115])
```

```python
df_new.head(30)
```

```python
from sklearn.model_selection import train_test_split
RSEED = 42

data = EmbeddingVectors
X_train, X_test, y_train, y_test = train_test_split(data, df_new['target'], test_size=0.2, random_state=RSEED)

print("train_x/train_y list details, to make sure it is of the right form:")
print(len(X_train))
print(X_train)
print(y_train[:5])
print(len(y_train))
```

```python
from sklearn.linear_model import LogisticRegression

def fit(train_x,train_y):
    model = LogisticRegression()

    try:
        model.fit(train_x, train_y)
    except:
        pass
    return model

model = fit(X_train,y_train)
```

```python
predicted_labels = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc_score = accuracy_score(y_test, predicted_labels)

print("The logistic regression accuracy score is::")
print(acc_score)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)
pca_ = PCA(0.99, random_state=RSEED)
X_pca=pca_.fit_transform(X_scaled)
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, df_new['target'], test_size=0.2, random_state=RSEED)
model = fit(X_train_pca,y_train_pca)

predicted_labels_pca = model.predict(X_test_pca)
acc_score_pca = accuracy_score(y_test_pca, predicted_labels_pca)

print("The logistic regression accuracy score is::")
print(acc_score_pca)
```
