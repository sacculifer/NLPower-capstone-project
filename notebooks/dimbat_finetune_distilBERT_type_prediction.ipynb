```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle("../data/preprocess_train_dimbat_label_2013.pkl")
df.head(5)
```

```python
df.info()
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification, pipeline
from sklearn.metrics import confusion_matrix, classification_report
```

```python
tweets = list(df["text_clean"])
```

```python
#tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
#tf_model = TFAutoModelForSequenceClassification.from_pretrained("../models/dimbat_disaster_type_distilbert_model")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("sacculifer/dimbat_disaster_type_distilbert")

tf_model = TFAutoModelForSequenceClassification.from_pretrained("sacculifer/dimbat_disaster_type_distilbert")
```

```python
classifier = pipeline("text-classification", tokenizer=tokenizer, model=tf_model)
```

```python
output = classifier(tweets)
```

```python
output
```

```python
y_pred = [d["label"] for d in output]
```

```python
y_score = [d["score"] for d in output]
```

```python
df['predict'] = y_pred
```

```python
df['predict_score'] = y_score
```

```python
df['predict'] = df['predict'].replace('LABEL_1', 1)
df['predict'] = df['predict'].replace('LABEL_2', 2)
df['predict'] = df['predict'].replace('LABEL_3', 3)
df['predict'] = df['predict'].replace('LABEL_4', 4)
df['predict'] = df['predict'].replace('LABEL_5', 5)
df['predict'] = df['predict'].replace('LABEL_6', 6)
df['predict'] = df['predict'].replace('LABEL_7', 7)
df['predict'] = df['predict'].replace('LABEL_8', 8)
df['predict'] = df['predict'].replace('LABEL_9', 9)
df['predict'] = df['predict'].replace('LABEL_0', 10)
```

```python
df['p_label'] = y_pred
```

```python
df['p_label'].value_counts()
```

```python
df['relevance'].value_counts()
```

```python
df['predict'].value_counts()
```

```python
print(confusion_matrix(df["relevance"], df["predict"]))
print(classification_report(df["relevance"], df["predict"]))
```
